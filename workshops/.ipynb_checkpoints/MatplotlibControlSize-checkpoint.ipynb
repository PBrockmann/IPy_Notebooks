```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)
y = x ** 2

plt.figure(figsize=(4,4))

plt.plot(x, y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title sdsdf')
plt.savefig('test1.png', dpi=100)
plt.show()
```

```python
%matplotlib inline

import netCDF4
f=netCDF4.Dataset("/prodigfs/esg/CMIP5/merge/IPSL/IPSL-CM5A-LR/rcp85/mon/atmos/Amon/r1i1p1/latest/tas/tas_Amon_IPSL-CM5A-LR_rcp85_r1i1p1_200601-230012.nc")

lat = f.variables['lat'][:]
lon = f.variables['lon'][:]
var = f.variables['tas'][10]

from mpl_toolkits.basemap import Basemap, shiftgrid, addcyclic
import matplotlib.pyplot as plt
import numpy as np
 
plt.figure(figsize=(12,12))

# set up orthographic map projection with
# perspective of satellite looking down at 50N, 100W.
# use low resolution coastlines.
#map = Basemap(projection='ortho', lat_0=50, lon_0=0, resolution='l')
map = Basemap(projection='robin', lon_0=0, llcrnrlat=-90, urcrnrlat=90,\
            llcrnrlon=0, urcrnrlon=360, resolution='c')

# draw coastlines, country boundaries, fill continents.
map.drawcoastlines(linewidth=0.25)
map.drawcountries(linewidth=0.25)

# draw lat/lon grid lines every 30 degrees.
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

# shift data so lons go from -180 to 180
var, lon = shiftgrid(180., var, lon, start=False)
var, lon = addcyclic(var, lon)
lons, lats = np.meshgrid(lon, lat)
x, y = map(lons, lats)

map.contourf(x, y, var, 20)
#map.pcolor(x, y, var)

plt.colorbar(orientation='horizontal')
plt.title("A matplolib example")
plt.savefig('test2.png',dpi=100)
plt.show()
```
