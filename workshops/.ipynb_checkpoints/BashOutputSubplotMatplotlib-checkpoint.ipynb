##### Draw a map

```python
%matplotlib inline

import numpy
lat = f.variables['lat'][:]
lon = f.variables['lon'][:]
var = f.variables['tas'][6]

from mpl_toolkits.basemap import Basemap, shiftgrid, addcyclic
import matplotlib.pyplot as plt
import numpy as np
 
plt.figure(figsize=(10,10))

# set up orthographic map projection with
# perspective of satellite looking down at 50N, 100W.
# use low resolution coastlines.
#map = Basemap(projection='ortho', lat_0=50, lon_0=0, resolution='l')
map = Basemap(projection='robin', lon_0=0, llcrnrlat=-90, urcrnrlat=90, llcrnrlon=0, urcrnrlon=360, resolution='c')

# draw coastlines, country boundaries, fill continents.
map.drawcoastlines(linewidth=0.25)
map.drawcountries(linewidth=0.25)

# draw lat/lon grid lines every 30 degrees.
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

# shift data so lons go from -180 to 180
var, lon = shiftgrid(180., var, lon, start=False)
var, lon = addcyclic(var, lon)
lons, lats = numpy.meshgrid(lon, lat)
x, y = map(lons, lats)

map.contourf(x, y, var-273.15, 20, extend='both', cmap=plt.cm.spectral)
#map.pcolor(x, y, var-273.15)

plt.colorbar(orientation='horizontal')
#plt.title("title")
#plt.savefig('test3png.png',dpi=100)
plt.show()
```

##### Investigate subplot mechanism

##### PLot the 3 different IPSL configurations

```python
import os

fig = plt.figure(figsize=(16,14))

map = Basemap(projection='robin', lon_0=0, llcrnrlat=-90, urcrnrlat=90, llcrnrlon=0, urcrnrlon=360, resolution='c')

for n in range(0,len(listfiles)):
    f=netCDF4.Dataset(listfiles[n])
    lat = f.variables['lat'][:]
    lon = f.variables['lon'][:]
    var = f.variables['tas'][6]-273.15  
    
    plt.subplot(3,1,n)
    map.drawcoastlines(linewidth=0.25)
    map.drawcountries(linewidth=0.25)
    map.drawmeridians(np.arange(0, 360, 30))
    map.drawparallels(np.arange(-90, 90, 30))

    var, lon = shiftgrid(180., var, lon, start=False)
    var, lon = addcyclic(var, lon)
    lons, lats = numpy.meshgrid(lon, lat)
    x, y = map(lons, lats)

    cs = map.contourf(x, y, var, levels=range(-40,40,2), extend='both', cmap=plt.cm.spectral)
    #map.pcolor(x, y, var)
    map.colorbar(cs)
    
    plt.title(os.path.basename(listfiles[n]).split("_")[2])
    
```
